```python
from contrastive_experiment.utils import get_base_folder
from contrastive_experiment.workflows import TrainWorkflow
```

```python
TEXT_COL = 'preprocessedV2Text'
base_folder = get_base_folder()
model_dir = base_folder / 'contrastive-experiment/models/paraphrase_multilingual_mpnet'
output_dir = model_dir / 'positive_fixed'

config_parameters = {
    'num_samples': None,
    'pairing_strategy': 'positive_contrastive',
    'loss': 'contrastive',
    'model_name': 'sentence-transformers/paraphrase-multilingual-mpnet-base-v2',
    'num_train_epochs': 1,
    'per_device_train_batch_size': 32,
    'per_device_eval_batch_size': 32,
    'learning_rate': 2e-5,
    'warmup_ratio': 0.1,
    # 'fp16': True,  # Set to False if GPU can't handle FP16
    'bf16': False,  # Set to True if GPU supports BF16
    'eval_strategy': "steps",
    'eval_steps': 100,
    'save_strategy': "steps",
    'save_steps': 100,
    'logging_steps': 100,
    'run_name': 'contrastive-positive-pairs-fix_samples'

}
workflow = TrainWorkflow(base_folder, TEXT_COL, output_dir, config_parameters)

pre_results, post_results = workflow.run()
```
